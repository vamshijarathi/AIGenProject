```python
import numpy as np
```

```python
arr1 = np.array([10,20,30])
arr2 = np.array([1,2,3])
```

```python
print(arr1)
```

```python
# elements wise addition
sum_arr = arr1 + arr2 
print(sum_arr)
```

```python
# element-wise multiplication
product_arr = arr1 * 2
print(product_arr)
```
