### Exception Handling in Python

```python
def even_or_odd(number):
    try:
        if number % 2 == 0:
            return "Even"
        else:
            return "Odd."
    except TypeError:
        return "Invalid input. Please provide an integer."    
```

```python
even_or_odd(5)
```

```python
even_or_odd("5")
```

### Class and OOP
- class LLM : blueprint or building plan
- __init__ : constructor or building material
- self: refers as instance itself or which building we are referring to
- summarize : function and serves specific task or performs certain action on the building
- gemini = LLM("Gemini Pro"): object  or phycial building
- gemini.summarize : Method calling  or performing certain action on building

```python
class LLM:
    def __init__(self, model_name):
        self.model_name = model_name

    def summarize(self, text):
        return f"Summarized text by {self.model_name}: {text}"   
     
```

```python
gemini = LLM("Gemini Pro")
print(gemini.summarize("This is a sample text to be summarized."))
```

```python
class BankAccount:
    """
    Blueprint for creating individual bank accounts
    Each account has a name and its own unique balance
    """
    #constructor
    def __init__(self, owner_name, initial_balance=0):
        self.owner_name = owner_name
        self.balance = initial_balance
        print(f"Account for {self.owner_name} created with balance {self.balance}.")

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited {amount}. New balance is {self.balance}.")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds for withdrawal.")
        else:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance is {self.balance}.")    
                     

    def get_balance(self):
        return f"current balance is {self.balance} for {self.owner_name}."
```

```python
account_alice = BankAccount("Alice", 500)
account_bob = BankAccount("Bob")

print("-" * 20)

account_alice.deposit(150)
account_bob.deposit(50)

print("-" * 20)

account_alice.withdraw(100)
account_bob.withdraw(200)

print("-" * 20)

print(account_alice.get_balance())
print(account_bob.get_balance())    
```
